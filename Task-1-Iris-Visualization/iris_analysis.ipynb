# 🌸 Iris Dataset Visualization

## 📌 Objective
This project explores and visualizes the Iris dataset to understand patterns and relationships.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## 📊 Dataset Overview
We check the shape and column names of the dataset.

```python
print("Shape:", iris.shape)
print("Columns:", iris.columns)
```

## 🔍 Dataset Information
This shows data types and missing values.

```python
iris.info()
```

## 📈 Statistical Summary
Shows mean, min, max and other values.

```python
iris.describe()
```

## 🌼 Scatter Plot
Relationship between sepal length and petal length.

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

## 📊 Histogram
Distribution of features.

```python
iris.hist()
plt.show()
```

## 📦 Box Plot
Used to check outliers.

```python
sns.boxplot(data=iris)
plt.show()
```

## 📌 Insights

- Dataset contains 150 rows and 5 columns  
- No missing values  
- Petal length clearly separates species  
- Setosa is easy to identify  
- Some outliers exist
